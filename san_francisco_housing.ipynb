# Housing Rental Analysis for San Francisco

```python
# Importing the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

## Import the data 

```python
# Using the read_csv function and Path module, creating a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(Path("./Resources/sfo_neighborhoods_census_data.csv"))

# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

---

## Calculating and Ploting the Housing Units per Year

### Step 1: Using the `groupby` function to group the data by year. Aggregating the results by the `mean` of the groups.

```python
# Creating a numerical aggregation that groups the data by the year and then averages the results.
housing_units_by_year = sfo_data_df.groupby('year').mean()

# Reviewing the DataFrame
display(housing_units_by_year)
```

### Step 2: Using the `hvplot` function to plot the `housing_units_by_year` DataFrame as a bar chart. Making the x-axis represent the `year` and the y-axis represent the `housing_units`.

### Step 3: Styling and formating the line plot to ensure a professionally styled visualization.

```python
# Creating a visualizing aggregation explore the housing units by year
plot_housing_units_by_year=housing_units_by_year.hvplot.bar(
    x='year', 
    y='housing_units',
    xlabel='Year',
    ylabel='Housing Units',
    title="Housing Units in San Francisco from 2010 to 2016",
    color='blue',
    hover_color='orange',
    yformatter='%.0f'
    )

plot_housing_units_by_year
```

### Step 5: Question and answer:

**Question:** What is the overall trend in housing_units over the period being analyzed?

**Answer:** The overall trend seems to be quite flat which means that the year-over-year increase in housing units is small.

## Calculating and Plotting the Average Sale Prices per Square Foot

### Step 1: Grouping the data by year, and then averaging the results.

```python
# Creating a numerical aggregation by grouping the data by year and averaging the results
prices_square_foot_by_year = sfo_data_df.groupby('year').mean()
# Reviewing the resulting DataFrame
prices_square_foot_by_year
```

**Question:** What is the lowest gross rent reported for the years included in the DataFrame?

**Answer:** The lowest gross rent is $1239 that occured in 2010.

### Step 2: Creating a new DataFrame named `prices_square_foot_by_year` by filtering out the “housing_units” column. The new DataFrame includes the averages per year for only the sale price per square foot and the gross rent.

```python
# Filtering out the housing_units column, creating a new DataFrame 
# Keeping only sale_price_sqr_foot and gross_rent averages per year
prices_square_foot_by_year = sfo_data_df.groupby('year').mean().drop(columns=['housing_units'])

# Reviewing the DataFrame
prices_square_foot_by_year
```

### Step 3: Using hvPlot to plot the `prices_square_foot_by_year` DataFrame as a line plot.

> **Hint** This single plot will include lines for both `sale_price_sqr_foot` and `gross_rent`

### Step 4: Styling and formating the line plot to ensure a professionally styled visualization.


```python
# Plotting prices_square_foot_by_year. 
# Including labels for the x- and y-axes, and a title.
prices_square_foot_by_year.hvplot(
    xlabel='Year',
    ylabel='Gross Rent/Sale Price per Square Foot',
    title='Sale Price per Square Foot and Average Gross Rent - 2010-2016 - San Francisco',
    legend=True
    )
```

### Step 6: Using both the `prices_square_foot_by_year` DataFrame and interactive plots to answer the following questions:

**Question:** Did any year experience a drop in the average sale price per square foot compared to the previous year?

**Answer:** The average sale price droped slightly in 2011 but went up in 2012 and has continued the increasing trend since than.

**Question:** If so, did the gross rent increase or decrease during that year?

**Answer:** In 2011 the gross rent increased despite the drop in average sale price.

## Comparing the Average Sale Prices by Neighborhood

### Step 1: Creating a new DataFrame that groups the original DataFrame by year and neighborhood. Aggregating the results by the `mean` of the groups.

```python
# Grouping by year and neighborhood and then creating a new dataframe of the mean values
prices_by_year_by_neighborhood = sfo_data_df.groupby(['year','neighborhood']).mean()

# Reviewing the DataFrame
prices_by_year_by_neighborhood
```

### Step 2: Filtering out the “housing_units” column to create a DataFrame that includes only the `sale_price_sqr_foot` and `gross_rent` averages per year.

```python
# Filtering out the housing_units
prices_by_year_by_neighborhood = prices_by_year_by_neighborhood.drop(columns=["housing_units"])

# Reviewing the first and last five rows of the DataFrame
display(prices_by_year_by_neighborhood.head())
display(prices_by_year_by_neighborhood.tail())
```

### Step 3: Creating an interactive line plot with hvPlot that visualizes both `sale_price_sqr_foot` and `gross_rent`. Setting the x-axis parameter to the year (`x="year"`). Using the `groupby` parameter to create an interactive widget for `neighborhood`.

### Step 4: Styling and formating the line plot to ensure a professionally styled visualization.

```python
# Using hvplot to create an interactive line plot of the average price per square foot
# The plot has a dropdown selector for the neighborhood
prices_by_year_by_neighborhood.hvplot(
    groupby='neighborhood',
    xlabel='Year',
    ylabel='Gross Rent/Sale Price per Suare Foot',
    title='Sale Price per Square Foot and Gross Rent - 2010-2016 - by Neighborhood',
    legend=True)
```

### Step 6: Using the interactive visualization to answer the following question:

**Question:** For the Anza Vista neighborhood, is the average sale price per square foot for 2016 more or less than the price that’s listed for 2012? 

**Answer:** The average sale price per square foot for Anza Vista neighborhood in 2016 is significantly lower than it was in 2012. Interesting is that gross rent continues to grow which signalizes an oppornunity to buy and rent out.

## Building an Interactive Neighborhood Map

### Step 1: Reading the `neighborhood_coordinates.csv` file from the `Resources` folder into the notebook, and creating a DataFrame named `neighborhood_locations_df`. Setting the `index_col` of the DataFrame as “Neighborhood”.

```python
# Loading neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(
    Path("./Resources/neighborhoods_coordinates.csv"),
    index_col="Neighborhood")

# Reviewing the DataFrame
neighborhood_locations_df
```

### Step 2: Using the original `sfo_data_df` Dataframe, creating a DataFrame named `all_neighborhood_info_df` that groups the data by neighborhood. Aggregating the results by the `mean` of the group.

```python
# Calculating the mean values for each neighborhood
all_neighborhood_info_df = sfo_data_df.groupby("neighborhood").mean()

# Reviewing the resulting DataFrame
all_neighborhood_info_df
```

### Step 3: Reviewing the two code cells that concatenate the `neighborhood_locations_df` DataFrame with the `all_neighborhood_info_df` DataFrame. 

Note that the first cell uses the [Pandas concat function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html) to create a DataFrame named `all_neighborhoods_df`. 

The second cell cleans the data and sets the “Neighborhood” column. 

Be sure to run these cells to create the `all_neighborhoods_df` DataFrame, which you’ll need to create the geospatial visualization.

```python
# Using the Pandas `concat` function, join the 
# neighborhood_locations_df and the all_neighborhood_info_df DataFrame
# The axis of the concatenation is "columns".
# The concat function will automatially combine columns with
# identical information, while keeping the additional columns.
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
all_neighborhoods_df = all_neighborhoods_df.reset_index().dropna()

# Rename the "index" column as "Neighborhood" for use in the Visualization
all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"})

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

### Step 4: Using hvPlot with GeoViews enabled, creating a `points` plot for the `all_neighborhoods_df` DataFrame. Doing following:

* Set the `geo` parameter to True.
* Set the `size` parameter to “sale_price_sqr_foot”.
* Set the `color` parameter to “gross_rent”.
* Set the `frame_width` parameter to 700.
* Set the `frame_height` parameter to 500.
* Include a descriptive title.

```python
# Creating a plot to analyze neighborhood info
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size='sale_price_sqr_foot',
    color='gross_rent',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title='Average Gross Rent & Sale Price per Square Foot'
)
```

### Step 5: Using the interactive map to answer the following question:

**Question:** Which neighborhood has the highest gross rent, and which has the highest sale price per square foot?

**Answer:** By looking at the map, we can identify the largest circle that represents the **highest sale price per square foot** and that is **Union Square District**.

The neighborhood with the **highest gross rent** is **Westwood Park** with $3.959 monthly and that is shown by the darkest circle in the map.
